```python
from sklearn.feature_extraction.text import CountVectorizer
from catboost import CatBoostRegressor, Pool
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
```

```python
data = pd.read_csv('./train.csv')
data_test = pd.read_csv('./test.csv')
```

```python
X_train, X_valid, y_train, y_valid = train_test_split(data.drop(columns='Rating'), data['Rating'])
```

```python
bag = CountVectorizer(max_features=1000)
X_train = bag.fit_transform(X_train['Review'])
X_valid = bag.transform(X_valid['Review'])
```

```python
cat = CatBoostRegressor()
cat.fit(X_train, y_train, eval_set=Pool(X_valid, y_valid), )
```

# Prediction

```python
X_test = bag.transform(data_test['Review'])
```

```python
np.savetxt('./submission.csv', cat.predict(X_test), fmt='%1.3f')
```

# Testing

```python
ground_truth = np.loadtxt('./answers.csv')
preds = np.loadtxt('./submission.csv')
r2_score(ground_truth, preds)
```
